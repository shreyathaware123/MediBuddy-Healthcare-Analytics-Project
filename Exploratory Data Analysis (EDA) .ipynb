# 5. Does a study of persons BMI get the company any idea for the insurance claim that it would extend?


```python
# 1. Checking correlation between BMI and Charges

correlation = df['BMI'].corr(df['Charges in INR'])
print(f"Correlation between BMI and Charges is = {correlation:.2f}")

# 2. Scatter plot of BMI vs Charges
# -----------------------------
plt.figure(figsize=(6,5))
plt.scatter(df['BMI'], df['Charges in INR'], alpha=0.5, color='blue')
plt.title('BMI vs Charges')
plt.xlabel('BMI')
plt.ylabel('Charges in INR')
plt.grid(True)
plt.show()

# 3. Categorize BMI

def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['BMI'].apply(bmi_category)

# 4. Calculate average charges per BMI category
avg_charges = df.groupby('BMI_Category')['Charges in INR'].mean()

# 5. Bar plot of average charges per BMI category

plt.figure(figsize=(6, 5))
bars = plt.bar(avg_charges.index, avg_charges.values, color=['skyblue','green','orange','red'])
plt.title('Average Charges by BMI Category')
plt.ylabel('Average Charges in INR')

# Adding labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 195, f'{yval:.0f}', ha='center', va='bottom')  # 500 is padding

plt.show()
```

## 9. High-Cost Claim Share by BMI Category

```python
# Defining high-cost claims (top 25%)
high_cost_threshold = df['Charges in INR'].quantile(0.75)
df['High_Cost'] = df['Charges in INR'] > high_cost_threshold

# Calculating high-cost claim share by BMI category
high_cost_bmi = df.groupby('BMI_Category')['High_Cost'].mean() * 100

# Plot
plt.figure(figsize=(6,3))
plt.bar(high_cost_bmi.index, high_cost_bmi.values, color='red')

plt.title('High-Cost Claim Percentage by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('High-Cost Claims (%)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

#### Business Insight

## 10. Claim Volatility by Age Group

```python
# Grouping by age bins
age_bins = [18, 25, 35, 45, 55, 65, 100]
age_labels = ['18-24','25-34','35-44','45-54','55-64','65+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Calculating standard deviation of charges per age group
volatility_by_age = df.groupby('Age_Group')['Charges in INR'].std()

# Plot
plt.figure(figsize=(6,3))
plt.bar(volatility_by_age.index, volatility_by_age.values, color='orange')

plt.title('Claim Volatility (Std Dev) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Claim Volatility (INR)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
